# Laad de beelden

```python
beeld_data = os.listdir(pad_naar_beelden)
```

```python
tot_start_time = datetime.now()

for beeld in beeld_data:
    
    #timer start
    start_time = datetime.now()
    
    img_path = os.path.join(os.path.join(pad_naar_beelden), beeld) # huidige beeldpad maken 
    
    img = cv2.imread(img_path) # laad het beeld met cv2.imread
    
    img = cv2.resize(img, dsize=(5000,5000), interpolation=cv2.INTER_CUBIC)
    
#     print(beeld[:-4])
    try:
        os.mkdir(nieuwe_pad_naar_beelden + '\\' + beeld[:-4] ) # maak folder
    except OSError as error:
        print(error) # error output
    
    full_path = os.path.join(nieuwe_pad_naar_beelden, beeld[:-4])
#     full_path = os.path.join(nieuwe_pad_naar_beelden + '\\' + beeld)
    
    hoogte = img.shape[0]
    breedte = img.shape[1]
    
    hoog = hoogte // 25
    breed = breedte // 25
    
    for x in range(0,hoogte, hoog):
        for y in range(0,breedte, breed):
        
            beeld_gesplitst = img[x:x+hoog, y:y+breed]
        
    #         full_path = os.path.join(nieuwe_beeld_locatie, beeld)
        
            cv2.imwrite(full_path + '/' + beeld[:13] + '_' + str(y) + '_' + str(x) + '.png',beeld_gesplitst)
            
    #time stop
    end_time = datetime.now()
    print(full_path + ' is gemaakt in ' + str(end_time - start_time) + ' seconden')
    
#     #TIPS: Originele file verwijderen
#     if os.path.exists(img_path):
#         print(img_path + ' is verwijderd')
#         os.remove(img_path) # een file verwijderen
#     else:
#         print('De file bestaat niet!')
    
# Totaal process in seconden
tot_end_time = datetime.now()
print( 'Totaal in ' + str(tot_end_time - tot_start_time) + ' seconden')
```

## Map voorbereiden
Verplaatsen alle beeld met hetzelfde id naar een map

```python
# locatie initializeren
beelden_map = r'C:\Users\agusb\Desktop\beeld_data\test_beeld_gesplitst'
nieuwe_locatie = r'C:\Users\agusb\Desktop\beeld_data\data2'
```

```python
beelden_map = r'C:\Users\agusb\Desktop\notebookp\verwerkt_data'
nieuwe_locatie = r'C:\Users\agusb\Desktop\beeld_data\data_gesplits_verzameld'
```

```python
dt = pd.read_csv('P2_Fabric_data_csv.csv', encoding = 'ISO-8859-1', skipinitialspace = True)
```

### Nieuwe map creeen

```python
try:
    os.mkdir(nieuwe_locatie) # maak folder
except OSError as error:
    print(error) # error output
```

### Laad map

```python
data_mappen = os.listdir(beelden_map)
```

```python
dt
```

### Vervang alle spatie in kolommen naam

```python
# Vervang alle spatie in de naam van kolommen naar '_'
dt.columns = [c.replace(' ', '_') for c in dt.columns]
```

```python
dt
```

### Drop Knit

```python
dt2 = dt[dt['Structure_and_Construction'] != 'Knit']
```

```python
dt2
```

## Maak submap

```python
# Haal ids
dt_ids = dt2['General_Identifier'] # haal kolom 'General Identifier' van csv

for idx in dt_ids: # loop door id's
    sub_data = os.path.join(nieuwe_locatie, idx) # initialiseer sub folder met id als de naam
    try:
        os.mkdir(sub_data) # maak sub-folder
    except OSError as error:
        print(error) # error
```

## Verplaats alle beelden

```python
for dt_map in data_mappen:
    dt_pad = os.path.join(beelden_map, dt_map)
    dt_submap = os.listdir(dt_pad)
    
    for idex in dt_ids: #loop door id's
        images = [f for f in dt_submap if idex in f] # verzamel beelden die behoren bij elk id
#         print(images)
        for image in images: #loop door beelden
            #create old and nieuw path
            new_path = nieuwe_locatie + '/'+ idex +'/'+ image # maak een path waar je de beelden naartoe wilt verplaatsen 
            old_path = beelden_map + '/' + dt_map +'/'+ image # maak een path waar je de beelden wilt halen
            shutil.move(old_path, new_path) # verplaats beelden naar "new_path"   
```
